```python
import pandas as pd
import plotly.express as px
import plotly.figure_factory as ff
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Student_Perfomance_Cleaned.csv')
#display first 5 rows and all datatypes.
data.head(), data.dtypes
```

```python
sns.set(style="whitegrid")
```

```python
# Creating an interactive box plot
fig = px.box(data, x='gender', y='math score', color='gender',
             title='Interactive Box Plot of Math Scores by Gender',
             labels={'gender': 'Gender', 'math score': 'Math Score'},
             points='all')  # Show all points
fig.show()
```

```python
# Calculating average scores by parental level of education
avg_scores = data.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean().reset_index()

# Creating an interactive bar plot
fig = px.bar(avg_scores, x='parental level of education',
              y=['math score', 'reading score', 'writing score'],
              title='Average Scores by Parental Level of Education',
              labels={'value': 'Average Score', 'parental level of education': 'Parental Level of Education'},
              barmode='group')
fig.show()
```

```python
fig = px.scatter(data, x='reading score', y='writing score',
                 color='gender', symbol='test preparation course',
                 title='Scatter Plot of Reading vs. Writing Scores',
                 labels={'reading score': 'Reading Score', 'writing score': 'Writing Score'},
                 hover_data=['gender', 'test preparation course'])
fig.show()
```
