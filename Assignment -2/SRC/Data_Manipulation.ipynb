```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("cleaned_train.csv")
```

```python
df.head()
```

**A. Selecting specific columns**

```python
selected_data = df[['Name', 'Year', 'Mileage', 'Price']]
selected_data.head()
```

**B. Filtering rows where manufacture year> 2015**

```python
filtered_data = df[df['Year'] > 2015]
filtered_data.head()
```

**C.Rename Columns**

```python
df = df.rename(columns={'Kilometers_Driven': 'Odometer', 'Owner_Type': 'Previous_Owners'})
print(df.columns)
```

**D. Arranging Data by Price in Ascending Order**

```python
arr_data = df.sort_values(by='Price', ascending=True)
arr_data.head()
```

**E.Summarize**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
descriptive_stats = df[['Price', 'Mileage', 'Car Age']].describe()
print(descriptive_stats)
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Price'], bins=30, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Car Prices', fontsize=16)
plt.xlabel('Price ($)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.axvline(df['Price'].mean(), color='red', linestyle='dashed', linewidth=1, label='Mean Price')
plt.axvline(df['Price'].median(), color='orange', linestyle='dashed', linewidth=1, label='Median Price')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Mileage'], bins=30, kde=True, color='lightgreen', edgecolor='black')
plt.title('Distribution of Mileage', fontsize=16)
plt.xlabel('Mileage (km)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.axvline(df['Mileage'].mean(), color='red', linestyle='dashed', linewidth=1, label='Mean Mileage')
plt.axvline(df['Mileage'].median(), color='orange', linestyle='dashed', linewidth=1, label='Median Mileage')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Car Age', y='Price', data=df, alpha=0.7, color='green')
plt.title('Price vs. Car Age', fontsize=16)
plt.xlabel('Car Age (Years)', fontsize=14)
plt.ylabel('Price in k ($)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xscale('linear')
plt.yscale('linear')
plt.show()

```
